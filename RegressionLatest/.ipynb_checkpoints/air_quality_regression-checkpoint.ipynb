## Data set Information

The dataset contains 9358 instances of hourly averaged responses from an array of 5 metal oxide chemical sensors embedded in an Air Quality Chemical Multisensor Device. The device was located on the field in a significantly polluted area, at road level,within an Italian city. Data were recorded from March 2004 to February 2005 (one year)representing the longest freely available recordings of on field deployed air quality chemical sensor devices responses. Ground Truth hourly averaged concentrations for CO, Non Metanic Hydrocarbons, Benzene, Total Nitrogen Oxides (NOx) and Nitrogen Dioxide (NO2) and were provided by a co-located reference certified analyzer. We used data to predict the amount of CO based on other parameters in this case Nitrogen dioxide(NO2) concentration since that is the most correlated one to the Carbon Monoxide(CO) content.

## Processing data

```python
air_data.dropna(axis=0, how='all')
```

## Features

```python
features = air_data
```

```python
features = features.drop('Date', axis=1)
features = features.drop('Time', axis=1)
features = features.drop('NO2(GT)', axis=1)
features = features.drop('PT08.S4(NO2)', axis=1)
```

```python
labels = air_data['NO2(GT)'].values
```

```python
features = features.values
```

## Train and test portions

```python
from sklearn.cross_validation import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=0.3)
```

```python
print("X_trian shape {}".format(X_train.shape))
print("y_train shape {}".format(y_train.shape))
print("X_test shape  {}".format(X_test.shape))
print("y_test shape  {}".format(y_test.shape))
```

### Linear Regression

```python
from sklearn.linear_model import LinearRegression
```

```python
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
print("Predicted values:", regressor.predict(X_test))
```

```python
print("R^2 score for liner regression: ", regressor.score(X_test, y_test))
```

```python
print('Coefficients: \n', regressor.coef_)
```

```python
print("Mean squared error: %.2f"
      % mean_squared_error(y_test, regressor.predict(X_test) ))
```

### Support Vector Regression

```python
from sklearn.cross_validation import KFold
from sklearn.svm import SVR
```

```python
support_regressor = SVR(kernel='rbf', C=1000)
support_regressor.fit(X_train, y_train)
```

```python
print("Coefficient of determination R^2 on train set: {}".format(support_regressor.score(X_train, y_train)))
```

```python
print("Coefficient of determination R^2 on test set: {}".format(support_regressor.score(X_test, y_test)))
```

### Decision tree regression

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
dtr = DecisionTreeRegressor()
dtr.fit(X_train, y_train)
```

```python
print("Coefficient of determination R^2 on train set: {}".format(dtr.score(X_train, y_train)))
```

```python
print("Coefficient of determination R^2 on test set: {}".format(dtr.score(X_test, y_test)))
```
